```python
# Add the preprocessing folder to the path
import sys
import os
import numpy as np
sys.path.append('../')
import json

os.getcwd()
```

```python
csdm_dataset_path = "../datasets/sw96_CSDM/features.dat"
csdm_metadata_path = "../datasets/sw96_CSDM/feature_metadata.json"
label_path = "../datasets/sw96_CSDM/labels.npy"


with open(csdm_metadata_path, "r") as meta_file:
    # metadata = json.load(meta_file)
    csdm_metadata = json.load(meta_file)

csdm_shape = tuple(csdm_metadata["shape"])
csdm_dtype = csdm_metadata["dtype"]

features = np.memmap(csdm_dataset_path, dtype=np.complex128, mode="r", shape = csdm_shape)
labels = np.load(label_path, mmap_mode="r")
```

```python
import matplotlib.pyplot as plt

csd_mat_mean = np.mean(features[0], axis = 0)

csd_mat_mean.shape

plt.figure(figsize=(10, 8))
plt.imshow(np.abs(csd_mat_mean), cmap='viridis', origin='lower')
plt.colorbar(label="CSD Magnitude")
# plt.title(f"CSD Matrix Magnitude at {selected_freq} Hz")
plt.xlabel("Signal Index")
plt.ylabel("Signal Index")
plt.show()
```
